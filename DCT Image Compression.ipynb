Importing Required Libraries

```python
from scipy.fftpack import dct, idct
from skimage.io import imread
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pylab as plt
```

DCT Functions with Scipy

```python
# implement 2D DCT
def dct2(a):
    return dct(dct(a.T, norm='ortho').T, norm='ortho')

# implement 2D IDCT
def idct2(a):
    return idct(idct(a.T, norm='ortho').T, norm='ortho')    
```

```python
im = rgb2gray(imread('terrain.jpg')) 

DCT =  dct2(im)
r_im = idct2(DCT)
```

```python
print("Image Shape :", np.shape(im))
print("DCT Shape :", np.shape(DCT))
```

```python
# plot original and reconstructed image
plt.gray()
fig, axs = plt.subplots(1, 2, figsize = (16,8))
axs[0].imshow(im)
axs[0].axis('off')
axs[0].set_title("Original Image")

axs[1].imshow(r_im)
axs[1].axis('off')
axs[1].set_title("Reconsturcted Image (100% Coefficients)")

plt.show()
```

```python
thresh = 0.0001
dct_thresh = DCT * (abs(DCT) > (thresh*np.max(DCT)))
P = round((np.sum(abs(DCT) > (thresh*np.max(DCT)))/im.size)*100, 2)
r_im = idct2(dct_thresh)

fig, axs = plt.subplots(1, 2, figsize = (16,8))
axs[0].imshow(im)
axs[0].axis('on')
axs[0].set_title("Original Image")

axs[1].imshow(r_im)
axs[1].axis('off')
title = "Reconsturcted Image ("+str(P)+"% Coefficients)"
axs[1].set_title(title)
plt.show()
```
